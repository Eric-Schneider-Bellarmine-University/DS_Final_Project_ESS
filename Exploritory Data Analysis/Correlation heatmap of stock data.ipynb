```python
import pandas
import seaborn
```

```python
stockdata=pandas.read_csv('Stock_Dataset.csv')
```

```python
stockdata=stockdata.drop(['Stock_Title'],axis=1,inplace=False)
```

```python
correlation_matrix=seaborn.heatmap(stockdata.corr())
```

```python
correlation_matrix.get_figure().savefig('Correlation Matrix')
```
