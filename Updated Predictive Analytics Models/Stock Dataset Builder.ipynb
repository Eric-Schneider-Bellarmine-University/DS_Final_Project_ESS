The current stock list is a near comprehensive list of all of the stocks sold on the New York Stock Exchange. To ensure the randomness of the sample, its best to randomly remove a set number of stocks. This ensures that the data is as much of a simple random sample as possible without compromising the size of the dataset. In this case, reducing the number of cases to 8000 seems appropriate.

```python
def random_set(dset,removal):
    '''takes the list of stocks and randomly removes a given amount in order to ensure a random sample'''
    for removals in range(0,removal):
        dset.remove(dset[random.choice(range(0,len(dset)))])
    return dset
```

Each Ratio is given its own method to calculate the individual amounts. The year input represents the binary amount to correspond to the statement amount of year 2023(0), or 2022(1). The index is how they appear on the financial statements.

```python
def get_roe(balance,income,year):
    if len(income)==0 or len(balance)==0:
        return None
    else:
        try:
            equity=balance['Stockholders Equity'][year]
        except:
            return 0
        else:
            equity=balance['Stockholders Equity'][year]
        net=income['Net Income'][year]
        return net/equity if equity!=0 else None
```

```python
def get_current(balance,year):
    if len(balance)==0:
        return None
    else:
        
        try:
            c_asset=balance['Current Assets'][year]
        except:
            c_asset=0
        else:
            c_asset=balance['Current Assets'][year]   
            
        try:
            c_liab=balance['Current Liabilities'][year]
        except:
            try:
                c_liab=balance['Current Debt'][year]
            except:
                c_liab=0
            else:
                c_liab=balance['Current Debt'][year]
        else:
            c_liab=balance['Current Liabilities'][year]
        return c_asset/c_liab if c_liab!=0 else 1
```

```python
def get_turnover(balance,income,year):
    if len(balance)==0:
        return None
    else:
        
        try:
            sales=income['Total Revenue'][year]
        except:
            return 1
        else:
            sales=income['Total Revenue'][year]
        try:    
            assets=balance['Total Assets'][year]
        except:
            assets=0
        return sales/assets if assets!=0 else None
```

```python
def get_wacc(balance,income,beta,year):
    if len(income)==0:
        return None
    else:
        risk_free_rate=.0438
        market_rate=.093
        
        try:
            tax_rate=income['Tax Provision'][year]/income['Pretax Income'][year]
        except:
            tax_rate=0
        else:
            tax_rate=income['Tax Provision'][year]/income['Pretax Income'][year]
        
        
        try:
            cap=balance['Total Capitalization'][year]
        except:
            cap=1
        else:
            cap=balance['Total Capitalization'][year]
        
        try:
            liab=balance['Total Debt'][year]
        except:
            liab=1
        else:
            liab=balance['Total Debt'][year]
        w_e=cap/(cap+liab)
        w_d=liab/(cap+liab)
        
#Not every company has interest expense on their income statements.
        try:
            interest=income['Interest Expense'][year]
        except:
            interest=0
        else:
            interest=income['Interest Expense'][year]
        
        try:
            avg_liab=(balance['Total Debt'][year]+balance['Total Debt'][year+1])/2
        except:
            avg_liab=0
        else:
            avg_liab=(balance['Total Debt'][year]+balance['Total Debt'][year+1])/2
            
        k_e=risk_free_rate+beta*(market_rate-risk_free_rate)
        k_d=(interest/avg_liab)*(1-tax_rate) if avg_liab!=0 else 1
        return w_e*k_e+w_d*k_d
```

```python
def get_div(income,cashflow,price,year):
    if len(price)==0:
        return None
    else:
        try:
            div=cashflow['Cash Dividends Paid'][year]
        except:
            return 0
        else:
            div=cashflow['Cash Dividends Paid'][year]
        try:
            shares=income['Basic Average Shares'][year]
        except:
            shares=0
        else:
            shares=income['Basic Average Shares'][year]

        end_price=price['Close'][-1]
        if shares==0 or end_price==0:
            return 0
        else:
            return ((div*-1)/shares)/end_price
```

```python
def get_earnings(income,price,year):
    if len(price)==0:
        return None
    else:
        try:
            eps=income['Basic EPS'][year]
        except:
            eps=0
        else:
            eps=income['Basic EPS'][year]
            
        end_price=price['Close'][-1]
        return eps/end_price if end_price!=0 else 0
```

```python
def get_rev(income,year):
    if len(income)==0:
        return None
    else:
        try:
            income['Total Revenue'][year]/income['Total Revenue'][year+1]
        except:
            return 1
        else:
            return income['Total Revenue'][year]/income['Total Revenue'][year+1]
```

```python
def get_ni(income,year):
    if len(income)==0:
        return None
    else:
        try:
            income['Net Income'][year]/income['Net Income'][year+1]
        except:
            return 1
        else:
            return income['Net Income'][year]/income['Net Income'][year+1]
```

```python
def get_price(price):
    if len(price)==0:
        return None
    else:
        start_price=price['Open'][0]
        end_price=price['Close'][0]
        return (end_price-start_price)/start_price if start_price!=0 else 1
```

```python
def identify_outliers(company,balance,income,cashflow,price2023,beta):
    try:
        if get_price(price2023)>1 or get_price(price2023)<-1:
            return True
    except:
        return True
    
    if len(balance)>2 and len(income)>2 and len(cashflow)>2:
        price2022=yfinance.Ticker(company).history(start='2022-01-01',end='2023-01-01')
        try:
            if get_price(price2022)>1 or get_price(price2022)<-1:
                return True
        except:
            return True
    
    return False
```

```python
roe=[]
current=[]
turnover=[]
wacc=[]
div=[]
earning=[]
rev=[]
ni=[]
price_change=[]
```

```python
def propogate(company):
    balance=yfinance.Ticker(company).balance_sheet.T
    if len(balance)==0:
        return None
    income=yfinance.Ticker(company).income_stmt.T
    cashflow=yfinance.Ticker(company).cash_flow.T
    price2023=yfinance.Ticker(company).history(start='2023-01-01',end='2024-01-01')
    try:
        beta=yfinance.Ticker(company).info['beta']
    except:
        beta=1
    else:
        beta=yfinance.Ticker(company).info['beta']
    
    if identify_outliers(company,balance,income,cashflow,price2023,beta):
        return None
        
    roe.append(get_roe(balance,income,0))
    current.append(get_current(balance,0))
    turnover.append(get_turnover(balance,income,0))
    wacc.append(get_wacc(balance,income,beta,0))
    div.append(get_div(income,cashflow,price2023,0))
    earning.append(get_earnings(income,price2023,0))
    rev.append(get_rev(income,0))
    ni.append(get_ni(income,0))
    price_change.append(get_price(price2023))
    if len(balance)>2 and len(income)>2 and len(cashflow)>2:
        price2022=yfinance.Ticker(company).history(start='2022-01-01',end='2023-01-01')
    
        roe.append(get_roe(balance,income,1))
        current.append(get_current(balance,1))
        turnover.append(get_turnover(balance,income,1))
        wacc.append(get_wacc(balance,income,beta,1))
        div.append(get_div(income,cashflow,price2022,1))
        earning.append(get_earnings(income,price2022,1))
        rev.append(get_rev(income,1))
        ni.append(get_ni(income,1))
        price_change.append(get_price(price2022))
    
for stocks in stocks_list[:]:
    print(str(int(stocks_list.index(stocks)+1))+". "+stocks)
    propogate(stocks)
```
